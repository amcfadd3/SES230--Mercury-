## 1\. Introduction

##### The hypothesis for the project is that earthquakes shallower than 100m are more likely to occur in primarily metamorphic rock, compared to regions with primarily igneous or sedimentary rock. This is based on the scientific question of what rock type are shallow earthquakes most likely to occur in? This hypothesis and scientific question are significant because it gives insight into how regions of similar lithology around a fault may function thus allowing us to predict and mitigate hazards of earthquakes of varying magnitudes.   
##### The scientific background begins with the review paper, *Displacement/Length Scaling Relationships for Normal Faults; a Review, Critique, and Revised Compilation* (Lathrop et al., 2022). This paper describes how, "The relationship between normal fault displacement (D) and length (L) varies due to numerous factors, including fault size, maturity, basin tectonic history, and host rock lithology."A database of 4059 normal faults were created then analyzed. The findings of this paper which are relevant include: faults containing stiffer rocks, igneous and carbonate rocks, tend to be under-displaced, and faults containing softer rocks, clastic sedimentary rocks, tend to be over-displaced. An interpretation of this data would be that normal faults with stiffer rocks have less displacement, or relative movement, when compared to normal faults with softer rocks. Another paper that was used for the initial scientific background is *The Influence of Lithology and Fault Source Volume on the Magnitude Frequency Distribution of Earthquakes* (Collettini & Tinti ,2025). This paper focuses on seismic data found in central Italy, using a b-value to compare and contrast rock deformation. The b-value as defined in the paper, The exact proportion of large and small earthquakes in a particular time or place is defined by the b-value of the earthquake Magnitude-Frequency-Distribution. This study provided information which confirmed that the lithology of a fault has an influence on the magnitude and frequency of earthquakes in a specific area, but the b-value was not further used in the final process of the project. The next relevant research paper is *An Experimental Investigation of the Effect of Lithology, Pore Pressure, and Stress on the Magnitude of Induced Seismicity Events* (Sathyanath & Sharma, 2025). In this paper they used a triaxial shear machine to record slip events, slip velocity, stick-slip behavior, and creep on multiple rock types. The following results were found: The friction coefficient of rocks with harder minerals were higher, rocks with higher amounts of quartz had higher magnitude stick-slip events, and harder minerals lead to overall more elastic deformation. The interpretation of these results is that rocks which contain harder minerals may lead to higher magnitude earthquakes.   
##### Further scientific background research was then done throughout the project in order to better understand the results that were found, including research on the mechanical properties of the upper crust. According to NoÃ«l et al. (2021), the deformation mechanisms of rocks in the upper crust are typically brittle and contain frictional deformation as opposed to plastic or ductile deformation. This information can give insight into how the earthquakes may be produced in depths shallower than 100m. More research on brittle deformation and its relation to earthquakes was found to be relevant as well. The paper *From slow to fast faulting: recent challenges in earthquake fault mechanics* (Nielsen S. 2017\) provides insight into how the types of strain which cause earthquakes are a mixture of brittle, plastic, and viscous deformational events. This paper allows us to see that brittle deformation is a part of earthquakes, which will be the type of deformation focused on due to the shallow nature of earthquakes being studied.

# 4\. Results

The analysis utilized a dataset of 13,055 earthquake points within California recorded between April 21, 2016, and April 21, 2026\. Each epicenter was spatially joined with the California Geologic Map to determine the underlying lithology.

### 4.1 Results

![Figure 5](figures/figures/California_Surface_Rock_Type_Percentages.png)

## 4.2 Description of Results

The final analysis of 13,055 shallow earthquakes in California reveals that sedimentary rock units are the primary host for seismic activity (*Fig. 5\)*, accounting for 34.7% of all events. When these are combined with metasedimentary units (24.9%), nearly 60% of the total recorded earthquakes occurred in sedimentary-based lithologies. In contrast, the data has shown the inverse of our hypothesis. Metamorphic rocks represent the least frequent host, containing only 0.9% of the recorded earthquake epicenters, and metaigneous rising to 6.5%, and finally, igneous rocks represented a significant 30.1%.

## 4.3 Evaluation of Analysis

What worked well: The use of GeoPandas for a spatial join between the USGS epicenter coordinates and the California Geologic Map was highly effective for categorizing large datasets. Filtering the data to include only shallow earthquakes (\<100m depth) and magnitudes of 2.0+ provided a more accurate correlation between surface lithology and seismic activity.

What did not work: Much of the initial plan did not work. The amount of surface area we intended to cover was far too large for the time frame we had, especially with revising our hypothesis so frequently. Narrowing it down from global (California, Japan, Italy), and filtered by mineral was a job far beyond our scope. After revising it to just US based locations (California, Alaska), we did not have the time needed to complete the Alaska map. This led to a necessary reduction in scope to focus solely on California.

Unexpected outcomes: The most significant unexpected outcome was the near-absence of earthquakes in metamorphic rock (0.9%), which contradicted our prediction that these "stiffer" basement rocks would be the most frequent site for shallow seismic nucleation.

## 4.4 Uncertainty and Significance

Sources of Uncertainty: The primary source of uncertainty is the 2.9% "Unknown" category, largely consisting of underwater epicenters where lithological data was unavailable. Furthermore, the analysis assumes that surface-mapped geology accurately represents the rock unit at the earthquake hypocenter, even at depths up to 100m.

Evaluation: Despite these factors, the results are highly meaningful. The 33.8% gap between the most frequent host (sedimentary) and the least frequent (metamorphic) is statistically significant and indicates a clear lithological preference for shallow seismic events in this region.

# 5\. Discussion

## 5.1 Hypothesis Evaluation

The project's hypothesis underwent four significant revisions as the team encountered technical challenges, computational limits, and new scientific data.

1. First Hypothesis: Initially, the team focused on a global scale (California, Japan, Italy), predicting that faults containing similar mineralogy or lithology would show a similarity in earthquake magnitude across those regions. This was shown to be quite unrealistic given how many variables and factors can change the mineralogy of the host rock.  
2. Second Hypothesis: The focus shifted to using the "b-value,â€ the proportion of small to large earthquakes which can be evaluated through the strength of a rock. However, filtering by specific minerals and calculating b-values globally was only effective in areas with thousands of data points, rendering it an ineffective tool.  
3. Third Hypothesis: The team simplified the scope to predicting that earthquake epicenters are more likely to occur in primarily metamorphic regions than in igneous or sedimentary regions. We revised our area of study to California and, optimistically, Alaska due to data accessibility.  
4. Fourth Hypothesis: To improve the correlation between surface geology and epicenter data, a final depth constraint was added: "Earthquakes within 100m of the surface are more likely in metamorphic units" This ensured the results were as accurate as possible by matching hypocenters to the specific surface geology recorded on national maps.

## 5.2 Scientific Interpretation  
The results identifying sedimentary rock units as the primary host for shallow earthquakes in California, can be interpreted through a combination of mineralogical composition and the mechanical properties of the upper crust (Johnson et al. 2017).  
A primary factor for this high seismic frequency is the presence of hard minerals within California's sedimentary sequences. According to Sathyanath & Sharma (2025), lithologies containing harder minerals, such as quartz, are associated with higher magnitude stick-slip events and greater elastic deformation. The geologic map of California confirms that the sedimentary units in the study area are predominantly sandstone, a rock type composed primarily of quartz and feldspar (Wu et al. 2013).This high quartz content likely increases the friction coefficient and rock strength, leading to the accumulation of higher stress levels that result in more frequent and detectable earthquake events.  
The concentration of earthquakes within the top 100m of the crust is further explained by the brittle nature of rocks at shallow depths (Johnson et al. 2017). Brittle deformation occurs when applied stress exceeds the internal strength of the rock, a process favored at high stress and the low pressure-temperature conditions characteristic of the upper crust (Brittle Deformation) . Unlike the deeper crust where rocks may exhibit ductile flow, the shallowest layers are prone to fracturing, creating the epicenters recorded in our dataset (Wu et al. 2013).  
Pore fluid pressure in sedimentary rocks are particularly susceptible to these brittle failures due to their high porosity and fluid content. Pore pressure is defined as the force exerted by fluids within the open spaces of the rock. As demonstrated by NoÃ«l et al. (2021), As pore fluid pressure builds up, it reduces the difference between total stress and fluid pressure, AKA â€œeffective confining pressure.â€ This then lubricates the fault planes and decreases the frictional forces that keep rock masses from sliding. In high fluid sedimentary settings, these pressurized fluids can overprint otherwise stable environments, forcing the rock into localized shear fracturing and earthquake-producing behavior (Lathrop et al. 2022).  
In summary, the combination of high quartz content in sandstones and the mechanical instability caused by pore fluid pressure makes shallow sedimentary rocks a more prolific host for near-surface seismicity than the denser, less porous metamorphic (0.9%) or metavolcanic (6.5%) units that initially formed our hypothesis.

# Reproducible Analysis Code

The following cells reproduce the spatial join and rock-type earthquake counts using files stored in this repository under data/raw/.




```python
from pathlib import Path

import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd


def find_project_root(start=None):
    """Find the repository root by looking for data/raw from the current notebook location."""
    start = Path.cwd().resolve() if start is None else Path(start).resolve()
    for path in [start, *start.parents]:
        if (path / "data" / "raw").exists():
            return path
    raise FileNotFoundError("Could not find a project root containing data/raw.")


PROJECT_ROOT = find_project_root()
RAW_DATA_DIR = PROJECT_ROOT / "data" / "raw"
FIGURES_DIR = PROJECT_ROOT / "figures"
FIGURES_DIR.mkdir(exist_ok=True)

ROCK_TYPES_PATH = RAW_DATA_DIR / "California_Rock_Types.shp"
EARTHQUAKES_PATH = RAW_DATA_DIR / "CA_EQ_Shallow.shp"

print(f"Project root: {PROJECT_ROOT}")
print(f"Rock types file exists: {ROCK_TYPES_PATH.exists()}")
print(f"Earthquakes file exists: {EARTHQUAKES_PATH.exists()}")
```

```python
rock_map = gpd.read_file(ROCK_TYPES_PATH)
quakes = gpd.read_file(EARTHQUAKES_PATH)

print("Shapefiles loaded successfully.")
print(f"Rock type polygons: {len(rock_map):,}")
print(f"Earthquake points: {len(quakes):,}")
print(f"Rock map CRS: {rock_map.crs}")
print(f"Earthquake CRS: {quakes.crs}")
```

```python
quakes = quakes.to_crs(rock_map.crs)

rock_type_column = "Rock_Type"
if rock_type_column not in rock_map.columns:
    raise KeyError(f"Expected '{rock_type_column}' in California_Rock_Types.shp. Available columns: {list(rock_map.columns)}")

rock_map_clean = rock_map[[rock_type_column, "geometry"]].copy()
joined = gpd.sjoin(quakes, rock_map_clean, how="inner", predicate="within")

joined_rock_column = "Rock_Type_right" if "Rock_Type_right" in joined.columns else rock_type_column
counts = joined[joined_rock_column].value_counts().sort_values(ascending=False)

summary = pd.DataFrame({
    "Rock Type": counts.index,
    "Count": counts.values,
    "Percentage": (counts.values / counts.sum() * 100).round(2),
})

print(f"Earthquakes matched to rock type polygons: {len(joined):,}")
print(f"Total earthquakes in input shapefile: {len(quakes):,}")
summary
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(summary["Rock Type"], summary["Count"], color="#4c78a8", edgecolor="black")
ax.set_title("Earthquake Count by Rock Type")
ax.set_xlabel("Rock Type")
ax.set_ylabel("Earthquake Count")
ax.tick_params(axis="x", rotation=35)
ax.grid(axis="y", alpha=0.3)
plt.tight_layout()

output_path = FIGURES_DIR / "notebook_earthquake_count_by_rock_type.png"
plt.savefig(output_path, dpi=300, bbox_inches="tight")
print(f"Saved figure to {output_path.relative_to(PROJECT_ROOT)}")
plt.show()
```
